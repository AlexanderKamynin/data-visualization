# Анализ числовых (вещественных) параметров без дефектов с помощью гистограмм

```python
# Функция для построения гистограмм
def hist_for_attribute_set(df, labels, ax, defect=None, bin_num=10, df2=pd.DataFrame(),
                           ylim=None, ylabel=None, y_scale='linear'):
    for i, j in enumerate(labels):
        for m, col in enumerate(j):
            ax[i][m].set_yscale(y_scale)
            ax[i][m].tick_params(labelsize=30)
            ax[i][m].set_xlabel(col, fontsize=30)

            # уберем лишние границы графика
            ax[i][m].spines['right'].set_visible(False)
            ax[i][m].spines['top'].set_visible(False)
            ax[i][m].grid(linestyle='--')

            ax[i][m].hist(df[col], bin_num, density=False, color='#377eb8', alpha=.75, )  # синий
            if not df2.empty:
                ax[i][m].hist(df2[col], bin_num, density=False, color='#e41a1c', alpha=.45)
                ax[i][m].legend(labels=['Норма', defect])  # красный '#e41a1c'

    # если графиков меньше чем есть в сетке, убираем лишние оси
    len_labels = len(labels)
    len_ax = len(ax)
    if len_labels < len_ax:
        for ax in ax[len_labels: len_labels + (len_ax - len_labels)]:
            ax.axis('off')
```

```python
labels = np.reshape(df51.columns, (5, 2))

fig, ax = plt.subplots(nrows=5, ncols=2, sharey=True, figsize=(40, 40))
# поэкспериментируйте с масштабом оси Y, параметр y_scale ('linear, 'symlog', 'log')
hist_for_attribute_set(df51, labels, ax, bin_num=20, y_scale='linear')

for i in range(5):
    ax[i][0].set_ylabel('Число записей', fontsize=30)
plt.suptitle('Гистограммы: параметры без дефектов', fontsize=30, y=.95)
fig.tight_layout(rect=[0, .03, 1, .95])
plt.savefig('df51_hist')
```

# Анализ числовых (вещественных) параметров без дефектов при помощи линейных графиков

```python
# разделение параметров, соответсвующих колебаниям и температуре
class DataFrame:
    def __init__(self, data_frame):
        self.data_frame = data_frame
        temp_columns = ["Temperature_in_the_left_bearing", "Temperature_in_the_right_bearing"]
        self.data_temp = data_frame[temp_columns]
        self.data_osc = data_frame.drop(columns=temp_columns)


df51_obj = DataFrame(df51)
df42_obj = DataFrame(df42)
df43_obj = DataFrame(df43)
df44_obj = DataFrame(df44)
```

```python
# линейные графики для множества параметров
def lineplot_for_atribute_set(df, df_x, labels, ax, df2 = False, ylim=None, ylabel=None):
    if df2:
        color='red'
    else:
        color = 'blue'
    for i, col in enumerate(labels):
        ax[i].set_xlabel(col, fontsize=20)
        ax[i].plot(df_x, df[col], color=color, linewidth=.65)
        ax[i].tick_params(labelsize=20)

        # уберем лишние границы графика
        ax[i].spines['right'].set_visible(False)
        ax[i].spines['top'].set_visible(False)
        ax[i].grid(linestyle='--')
```

```python
current_df = df51_obj
labels = current_df.data_osc.columns

fig, ax = plt.subplots(nrows=labels.size, ncols=1, sharey=True, figsize=(20, 40))

lineplot_for_atribute_set(df=current_df.data_osc, df_x=current_df.data_frame.index, labels=labels, ax=ax)
fig.suptitle('Изменение вещественных параметров при норме (колебания)', fontsize=30, y=1)
plt.savefig("df51_osc_linear")
plt.tight_layout()
plt.show()
```

```python
labels = current_df.data_temp.columns
fig, ax = plt.subplots(nrows=labels.size, ncols=1, sharey=True, figsize=(20, 40))

lineplot_for_atribute_set(df=current_df.data_temp, df_x=current_df.data_frame.index, labels=labels, ax=ax)
fig.suptitle('Изменение вещественных параметров при норме (температура)', fontsize=30, y=1)
plt.savefig("df51_temp_linear")
plt.tight_layout()
plt.show()
```

# Анализ числовых (вещественных) параметров с помощью матрицы корреляций

```python
# матрица корреляции намного симпатичнее выглядит в seaborn
def heatmap(df, ax, title):
  corr = df.corr()

  # пропустим пары со слабой корреляцией, попробуем поэкспериментировать с этой настройкой
  df_corr = corr[(corr>=.4) | (corr<=-.4)]

  # print(f'Empty columns: {get_empty_columns_names(df_corr)}')

  mask = np.triu(np.ones_like(corr))
  sns.heatmap(
    df_corr,
    cbar=False,
    mask=mask,
    ax=ax,
    annot=True,
    cmap='coolwarm',
    vmin=-1,
    vmax=1,
  )

  ax.title.set_text(title)
  ax.tick_params(
    top=False,
    bottom=False,
    left=False,
    right=False,
  )
```

```python
df51[['misalignment in the coupling', 'the presence of an imbalance', 'bushing with defects']] = np.full((df51.index.size,3), 0)

df42[['the presence of an imbalance', 'bushing with defects']] = np.full((df51.index.size, 2), 0)
df42['misalignment in the coupling'] = np.full(df51.index.size, 1)

df43[['misalignment in the coupling', 'bushing with defects']] = np.full((df51.index.size, 2), 0)
df43['the presence of an imbalance'] = np.full(df51.index.size, 1)

df44[['misalignment in the coupling', 'the presence of an imbalance']] = np.full((df51.index.size, 2), 0)
df44['bushing with defects'] = np.full(df51.index.size, 1)

df42_corr = pd.concat([df51, df42], axis=0)
df43_corr = pd.concat([df51, df43], axis=0)
df44_corr = pd.concat([df51, df44], axis=0)
```

## Несоосность в муфте

```python
current_defect = 'misalignment in the coupling'
columns = [ 'Fluctuations_X_in_the_left_bearing',
       'Fluctuations_Y_in_the_left_bearing',
       'Fluctuations_X_in_the_right_bearing',
       'Fluctuations_Y_in_the_right_bearing',
       'Fluctuations_X_in_the_left_bearing_smooth',
       'Fluctuations_Y_in_the_left_bearing_smooth',
       'Fluctuations_X_in_the_right_bearing_smooth',
       'Fluctuations_Y_in_the_right_bearing_smooth',
       'Temperature_in_the_left_bearing',
        'Temperature_in_the_right_bearing'
        , current_defect]


fig, ax = plt.subplots(figsize=(7, 7), constrained_layout=True)
hm = heatmap(df42_corr[columns], ax, "Матрица корреляции (несоосность в муфте)")
plt.savefig('df42_corr')
```
